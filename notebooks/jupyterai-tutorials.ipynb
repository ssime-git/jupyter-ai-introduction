## Data Science use case

Let's import some data and see how we can levarage this extension further.

```python
%%ai -f code
write the python code 
to load iris dataset from scikit learn 
into pandas dataframe
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable (species) to the DataFrame
df['target'] = iris.target

# Add the target names (species names) to the DataFrame
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Print the first few rows of the DataFrame
print(df.head())
```

## Data exploration

Let's one of the sample code provided.

```python
import matplotlib.pyplot as plt

# Plot a scatter plot to visualize the relationship between features
plt.figure(figsize=(8,6))
for species in df['species'].unique():
    species_df = df[df['species'] == species]
    plt.scatter(species_df[iris.feature_names], species_df['target'], label=species)
```

```python
# Corrected Scatter Plots - Feature vs. Target (for each feature)
plt.figure(figsize=(16, 10))  # Adjust figure size for multiple subplots

for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)  # Create subplots in a 2x2 grid
    for species in df['species'].unique():
        species_df = df[df['species'] == species]
        plt.scatter(species_df[feature], species_df['target'], label=species)
    plt.xlabel(feature)
    plt.ylabel('Species (Target)')
    plt.title(f'{feature} vs. Species')  # More informative title
    plt.legend()

plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()
```

```python
import seaborn as sns
# Box Plots - Distribution of Each Feature Grouped by Species
plt.figure(figsize=(16, 10)) # Adjust figure size

for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1) # Create subplots
    sns.boxplot(x='species', y=feature, data=df) # Use seaborn boxplot
    plt.title(f'{feature} by Species') # Title reflecting the feature
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout() # Adjust layout
plt.show()
```
